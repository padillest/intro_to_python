# Indexing and Slicing

Traversing through the container types

### Indexing

Calling an item in a container by its position

Denoted with a container type (list, tuple) and square brackets ([])

Format:

    my_list[index]

Python is zero-indexed, which allows for better traversals over container types

```python
app = ["tiktok", "twitter", "facebook",
       "instagram", "snapchat", "pinterest"]
```

```python
app[0] # returns tiktok 
```

```python
app[4] # returns snapchat
```

We can all reverse indices 

```python
app[-1]
```

```python
app[-4]
```

### Slicing 

Denoted by a container type (list, tuple) and square brackets ([])

Format: 

    my_list[start:stop:step]

It is left inclusive, right exclusive

What does that mean? 

my_list[0:2] returns the items at index 0 and 1, but not at 2

```python
app[0:4] # returns everything from index 0 (tiktok) TILL snapchat 
```

```python
# python likes to make assumptions 

app[2:] # python assumes you want the entire list if we don't specify the 'stop'
```

```python
app
```

```python
app[-4:-1] # start at index -4, and stop at index -1
```

You can only reverse the list when stepping in reverse

```python
# app[-1:-4:-1] 
# you must use steps 
```

#### Stepping

```python
app[::2] # returns every second item
```

```python
# we can also reverse the entire list

app[::-1]
```

### Arguments

Data outside of a function that can be passed to a function

```python
def greeting(*name):
    """
    A function that greets a user
    """
    for n in name: 
        print("welcome to the party, " + n)
    
greeting("dennis", "katie")
```

Used to reduce the amount of repetitive code when having to do the same thing over and over

```python
greeting(name = "maia")
```

### Arbitrary Arguments

When you don't know how many arguments will be passed

```python
# stores the arguments as a tuple 

# recall that a tuple is like a list, but the items cannot be changed

# here the '*' symbol indicates that we might have a varying amount 
# of arguments 

def day_care_attendance(*child_attendance):
    """
    A function to greet all the day care attendees
    """
    # loop through the tuple of children names
    for attendee in child_attendance: 
        # print hello with each of the attendees names 
        print("hello", attendee)
```

```python
day_care_attendance("dennis", "angelo")
```

### __init__

initializing properties of a class 

```python
class Student: 
    def __init__(self, name, yr):
        self.name = name 
        self.yr = yr
        
    def graduation_yr(self):
        print("you will graduate in {} years".format(4-self.yr))
```

```python
user1 = Student(name = "Dennis", 
                yr = 1)
```

```python
user1.name
```

```python
user1.yr
```

```python
user1.graduation_yr()
```

# Challenge Question 

Given a string, return a new string where "not " has been added to the front. However, if the string already begins with "not", return the string unchanged

Example:

    not_string('candy')  'not candy'
    
    not_string('x')  'not x'
    
    not_string('not bad')  'not bad'


```python
def not_string(str):
  not_str = "not"
  if (not_str in str[0:4]):
    return str 
  else: 
    return "not {}".format(str) 
```
